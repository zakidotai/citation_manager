```python
import bibtexparser
import os

import numpy as np
import pandas as pd

# with open('../krishnan.bib') as bibtex_file:
with open('../citations_zaki.bib', encoding='utf-8') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
```

```python
author_name = 'Zaki'
```

```python
papersdict = dict()
for b in bib_database.entries:
    papersdict[b['ID']] = b
```

```python
keys = []
for paper in bib_database.entries[::-1]:
    keys.extend(list(paper.keys()))

dfpapers = pd.DataFrame(columns=np.unique(keys))   
for paper in bib_database.entries[::-1]:   
    dfp = pd.DataFrame(paper, index=[0])
    dfpapers = pd.concat((dfpapers, dfp))

dfsorted = dfpapers.sort_values(by='year', ascending=False)
```

```python
start = len(bib_database.entries)
# for paper in bib_database.entries[::-1]:
for p in dfsorted.ID:
    paper = papersdict[p]
    # print(p.keys())
    os.makedirs(author_name, exist_ok = True)
    info = ['---\n','layout: paper\n','tagname: Krishnan\n']
    title_line = 'title: '+ '"' +list({paper['title']})[0] + '"\n' 
    info.append(title_line)
    info.append('cites: "0"\n')
    
    try:
        year_line = 'year: '+ '"' +list({paper['year']})[0] + '"\n' 
        info.append(year_line)
    except KeyError:
        year_line = 'year: '+ '"' +'To be updated' + '"\n' 
        info.append(year_line)
    author_line = 'author: '+ '"' +list({paper['author']})[0] + '"\n' 
    info.append(author_line)
    # pub_line = 'publisher: '+ '"' +list({paper['publisher']})[0] + '"\n' 
    # info.append(pub_line)
    try:
        journal_line = 'journal: '+ '"' +list({paper['journal']})[0] + '"\n' 
        info.append(journal_line)
    except:
        try:
            pub_line = 'publisher: '+ '"' +list({paper['publisher']})[0] + '"\n' 
            info.append(pub_line)
        except KeyError: 
            journal_line = 'journal: '+ '"' +'To be updated' + '"\n' 
            info.append(journal_line)
            pub_line = 'publisher: '+ '"' +'To be updated' + '"\n' 
            info.append(pub_line)
            
    info.append('---\n')
    info.append('<!--end excerpt-->\n')
    title=list({paper['title']})[0][:30]
    fname = f'./{author_name}/{start}_{title.strip()}.md'
    try:
        with open(fname, 'w') as f:
            f.write(''.join(info))
            start = start - 1
    except FileNotFoundError:
        title=list({paper['title']})[0][:30]
        title = ''.join(e for e in title if e.isalnum())
        fname = f'./{author_name}/{start}_{title.strip()}.md'
        
        with open(fname, 'w', encoding='utf-8') as f:
            f.write(''.join(info))
            start = start - 1
         
    # break
```
